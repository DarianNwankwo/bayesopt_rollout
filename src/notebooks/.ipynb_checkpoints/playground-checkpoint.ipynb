```julia
using Zygote
```

```julia
f(x) = 14x[1]^4 + 7x[1]*x[2]^2*sin(x[2]) - 3x[1]^2*x[3]^2*exp(x[3]*x[2])
```

```julia
Zygote.hessian(f, [1 1 1])
```

```julia
Zygote.hessian(x -> f(x), [1 1 1])
```

```julia
xk = rand(3)
for i = 1:12
    # Update xk
    xk = xk - (Zygote.hessian(f, xk) \ Zygote.gradient(f, xk)[1])
    println(f(xk))
end
```
